```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, mean_squared_error
```

```python
df = pd.read_csv('C:\\Users\\Hazel\\Desktop\\Data Analyst (2024)\\Power BI & ML\\car data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
#drop column name
df.drop('Car_Name',axis=1,inplace=True)
```

```python
#create dummy variables or one hot encoding
df=pd.get_dummies(df,columns=['Fuel_Type','Seller_Type','Transmission'],drop_first=True)
```

```python
df.head()
```

```python
#define our target
X=df.drop(['Selling_Price'],axis=1)
y=df['Selling_Price']
```

```python
#training and testing
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.7,random_state=123)
X_train.head()
```

```python
#preprocessing
scale=StandardScaler()
X_train_S=scale.fit_transform(X_train)
X_test_S=scale.fit_transform(X_test)
X_train_S
```

```python
#model
#model=LinearRegression()
#model.fit(X_train_S,y_train)---->0.62
model=ExtraTreesRegressor()
model.fit(X_train,y_train)
```

```python
#predict and test
y_pred=model.predict(X_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
cross_val_score(model, X_train, y_train, cv=10, scoring='r2').mean()
```

```python
#Overfitting 
-Good with the training data but poor with the testing data
-captures noises but not the pattern
#Underfitting(learning too litle)
-model is too simple to capture the real pattern
-fails ont the testing as well as training
```
